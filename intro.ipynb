## Using a Simple Linear regression model

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import sklearn
```

```python
df=pd.read_csv("salary_predict_dataset.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

fill test_score and interview_score with mean value 
***

```python
df['test_score'].fillna(df['test_score'].mean(),inplace=True)
df['interview_score'].fillna(df['interview_score'].mean(),inplace=True)
```

```python
df
```

```python
df['experience'].fillna(0,inplace=True)
```

```python
df
```

Converting experience data into integer
*******

```python
def string_to_number(w):
    dict={'zero':0,'one':1,'two':2,'three':3,'four':4,'five':5,'six':6,'seven':7,'eight':8,'nine':9,'ten':10,'eleven':11,'twelve':12,'thirteen':13,'fourteen':14,'fifteen':15,0:0}
    return dict[w]
```

```python
df['experience']=df['experience'].apply(lambda x:string_to_number(x))
```

```python
df
```

# Train model 


```python
X=df.iloc[:,:3]
Y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.1,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression

mymodel=LinearRegression()
mymodel.fit(xtrain,ytrain)
```

```python
ypred=mymodel.predict(xtest)
```

```python
mymodel.predict([[5,8,7]])
```
